```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.dpi']= 180
from matplotlib.colors import ListedColormap

def plotDailyTransactions(bank_statement):

    f = open(bank_statement)

    transactions = []
    days = []
    
    all_file = f.readlines()
    
    for tr in all_file[1:-1]:
        days.append(int(tr.split(',')[0]))
        transactions.append(float(tr.split(',')[1]))
    
    plt.plot(days, transactions, color='blue')
    plt.xlabel("Days")
    plt.ylabel('Transactions (USD)')
    plt.title('Monthly transactions')
    
    plt.show()
```

```python
plotDailyTransactions('bank_statement.csv')
```
